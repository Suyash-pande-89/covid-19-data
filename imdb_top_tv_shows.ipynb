#IMDB Top TV shows scrapper
This script is a simple code to help beginners to start there first scraping project.
Aim of this code is to create a dataframe with following variables:

1. Name of the TV show
2. Year it aired
3. Genre
4. IMDB Rating
5. Votes 





```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
!nvidia-smi
```

Following is the code for scraping

```python
# Get url and generate request
url = "https://www.imdb.com/search/title/?title_type=tv_series&num_votes=1000,&languages=en&sort=num_votes,desc"
response = requests.get(url)
soup = BeautifulSoup(response.content,'html.parser')

# Get all div tabs
movie_data = soup.findAll('div',attrs= {'class': 'lister-item mode-advanced'})

# Initialize as lists 
tvs = []
year = []
genre = []
rating = []
votes =[]

#create lists for the variables
for store in movie_data:
  tvs.append(store.h3.a.text)
  year.append(store.h3.find('span',class_ ="lister-item-year text-muted unbold").text.replace('(', '').replace(')', ''))
  genre.append(store.p.find('span',class_ ='genre').text.replace('\n','').replace('[','').replace(']',''))
  rating.append(store.find('div',class_="inline-block ratings-imdb-rating" ).text.replace('\n',''))
  votes.append(store.find_all('span', attrs = {'name': 'nv'})[0].text)

#create data frame
TVS_DF = pd.DataFrame({'Name of TV show': tvs, 'Year of relase': year, 'Genre': genre, 'Movie Rating': rating, 'Votes': votes})
```

```python
TVS_DF.sort_values(['Movie Rating'],ascending= False)
```
